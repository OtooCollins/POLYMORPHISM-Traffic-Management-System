```python
class TrafficDevice:            #Parent class

    def activate(self):          #activate method
        print("Traffic device activated")


class TrafficLight(TrafficDevice):   #child class traffic light
    def activate(self):
        print("Traffic Light: Green light ON")


class SpeedCamera(TrafficDevice):    #child class speed camera
    def activate(self):
        print("Speed Camera: Capturing speeding vehicles")


class PedestrianSignal(TrafficDevice):  #child class pedistrian signal
    def activate(self):
        print("Pedestrian Signal: Walk sign displayed")


class EmergencySiren(TrafficDevice):   #new child class
    def activate(self):
        print("Emergency Siren: Warning siren activated")

#objects
light = TrafficLight()
camera = SpeedCamera()
signal = PedestrianSignal()
siren = EmergencySiren()

devices = [light, camera, signal, siren]

#activate each device
for device in devices:
    device.activate()
```
